```
import pandas as pd
import matplotlib.pyplot as plt

fonte = "https://github.com/alura-cursos/imersao-dados-2-2020/blob/master/MICRODADOS_ENEM_2019_SAMPLE_43278.csv?raw=true"

dados = pd.read_csv(fonte)
dados.head()
```

```
dados_se = dados.query("SG_UF_RESIDENCIA == 'SE'")
dados_se.describe()
```

### 01 - Total de inscritos em SE



```
inscritos = len(dados_se)
print(f'Sergipe teve {inscritos} inscritos no ENEM 2019')
```

### 02 - Idade dos inscritos em SE

```
idade = dados_se["NU_IDADE"]
idade.plot.box(grid=True, figsize= (10,8))
plt.title('Idade dos inscritos em SE')

#Relacionar com dados nacionais
```

### 03 - Escola de origem

```
publica = dados_se.query('TP_ESCOLA == 2')["TP_ESCOLA"].value_counts().sort_index()
privada = dados_se.query('TP_ESCOLA == 3')["TP_ESCOLA"].value_counts().sort_index()

total_inscritos = [sum(publica), sum(privada)]

i = 0
proporcao = []
while i < len(total_inscritos):
  proporcao.append((total_inscritos[i] * 100) / inscritos)
  i = i + 1

df = pd.DataFrame(
    {
        'origem': proporcao
    },
    index = ['escola pública', 'escola privada']
)

df.plot.pie(title="Proporção dos alunos oriundos de escolas públicas e particulares", legend='none', autopct='%1.1f%%', y='origem', figsize=(8,8))

#Relacionar com dados nacionais
```

### 04 - Dados das notas em SE

```
provas = ["NU_NOTA_CN","NU_NOTA_CH","NU_NOTA_MT","NU_NOTA_LC","NU_NOTA_REDACAO"]

dados_se.describe()
dados_se[provas].boxplot(grid=True, figsize= (10,8))
plt.title('Notas por matérias em SE')

#Relacionar com dados nacionais
```
